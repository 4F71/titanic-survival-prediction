```python
#Veriyi Yükleme ve İlk Tanışma

#Gerekli kütüphaneleri projeye dahil etme
import pandas as pd
import numpy as np

# Veri setlerimizi okuma (artık aynı klasörde olduğu için anında bulunacaklar!)
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')

# Verinin doğru yüklendiğinden emin olmak için ilk 5 satırına bakalım
print("Eğitim Verisinin İlk 5 Satırı:")
print(train_data.head())

print("\n----------------------------------\n")

# Veri setinin genel yapısı ve eksik verileri görmek için özetini alalım
print("Eğitim Verisinin Genel Bilgileri:")
train_data.info()
```

```python

# --- Eksik Yaş Verilerini Doldurma ---
# Eksik yaşları, tüm yolcuların yaşlarının medyanı ile dolduruyoruz.
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
test_data['Age'].fillna(test_data['Age'].median(), inplace=True) # Aynı işlemi test verisine de yapmayı unutmuyoruz!

# ---Eksik Biniş Limanı Verilerini Doldurma ---
# Eksik 2 değeri, en sık kullanılan liman (mod) ile dolduruyoruz.
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)

# --- Kabin Sütununu Kaldırma ---
# Aşırı eksik veri içerdiği için Cabin sütununu kaldırıyoruz.
train_data = train_data.drop('Cabin', axis=1)
test_data = test_data.drop('Cabin', axis=1)

print("Veri temizleme işlemi başarıyla tamamlandı!")
```

```python
# Gerekli kütüphaneleri ve veri setlerini yükleyelim
import pandas as pd # Veri manipülasyonu için popüler bir kütüphane olan Pandas'ı içe aktarırız

# 'train.csv' ve 'test.csv' dosyalarını DataFrame olarak yükleriz
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')

# Bu, orijinal veri üzerinde çalışmak yerine yeni bir veri çerçevesi oluşturur
train_data = train_data.copy()
test_data = test_data.copy()

# 1. Eksik "Age" (Yaş) verilerini doldurma
# 'train_data' seti için 'Age' sütununun ortalama değerini hesaplarız
ortalama_yas_train = train_data['Age'].mean()
# 'Age' sütunundaki boş (NaN) değerleri, hesapladığımız ortalama ile doldururuz
train_data['Age'].fillna(ortalama_yas_train, inplace=True)

# 'test_data' seti için de aynı işlemi yaparız
ortalama_yas_test = test_data['Age'].mean()
test_data['Age'].fillna(ortalama_yas_test, inplace=True)

# 2. Eksik "Embarked" (Biniş Limanı) verilerini doldurma
# 'Embarked' sütununda en sık tekrar eden (mode) değeri buluruz
en_sik_liman = train_data['Embarked'].mode()[0]
# Eksik 'Embarked' verilerini, en sık tekrar eden liman ile doldururuz
train_data['Embarked'].fillna(en_sik_liman, inplace=True)

# 3. Gereksiz sütunları kaldırma
# 'Cabin', 'Name' ve 'Ticket' gibi model için anlamsız olan sütunları her iki veri setinden de sileriz
# 'axis=1' sütunları, 'inplace=True' ise işlemi kalıcı olarak uygular
train_data.drop(['Cabin', 'Name', 'Ticket'], axis=1, inplace=True)
test_data.drop(['Cabin', 'Name', 'Ticket'], axis=1, inplace=True)

# Şimdi eksik verileri tekrar kontrol edelim
print("Eksik veri kontrolü (train_data):")
# Eksik verilerin sayısını her sütun için tekrar yazdırarak veri temizliğinin başarılı olduğunu doğrularız
print(train_data.isnull().sum()) 
print("\nEksik veri kontrolü (test_data):")
print(test_data.isnull().sum())
```

```python
# 'train_data' veri setinin ilk 5 satırını görmek için head() komutunu kullanırız
# Bu, verinin genel bir görünümünü ve sütunların içeriğini anlamamızı sağlar
print("Eğitim Verisi (train_data) İlk 5 Satır:")
print(train_data.head())

# Bir sonraki satırda test verisinin ilk 5 satırını görüntüleriz
print("\nTest Verisi (test_data) İlk 5 Satır:")
print(test_data.head())

# ---

# Şimdi veri setlerinin genel bilgilerine bakalım
# 'info()' komutu, her sütunun veri tipini ve eksik olmayan değer sayısını gösterir
print("\nEğitim Verisi (train_data) Bilgileri:")
train_data.info()

print("\nTest Verisi (test_data) Bilgileri:")
test_data.info()
```

```python
# 'test_data' setindeki 'Fare' (Ücret) sütununun ortalama değerini hesaplarız
ortalama_ucret = test_data['Fare'].mean()

# 'Fare' sütunundaki eksik değeri (NaN) hesapladığımız ortalama ile doldururuz
test_data['Fare'].fillna(ortalama_ucret, inplace=True)

# Şimdi 'test_data' setindeki eksik veri kalıp kalmadığını kontrol edelim
print("Eksik veri kontrolü (test_data):")
print(test_data.isnull().sum())
```

```python
# 'FamilySize' (Aile Büyüklüğü) adında yeni bir sütun oluştururuz
# Bu sütun, SibSp ve Parch sütunlarının toplamı ve kişinin kendisi (+1) ile hesaplanır
train_data['FamilySize'] = train_data['SibSp'] + train_data['Parch'] + 1
test_data['FamilySize'] = test_data['SibSp'] + test_data['Parch'] + 1

# 'FamilySize' sütununu oluşturduktan sonra, 'SibSp' ve 'Parch' sütunlarını silebiliriz
# Çünkü yeni sütun, bu iki sütunun bilgisini zaten içeriyor
train_data.drop(['SibSp', 'Parch'], axis=1, inplace=True)
test_data.drop(['SibSp', 'Parch'], axis=1, inplace=True)

# Verinin son halini kontrol etmek için ilk 5 satırını yazdırırız
print("Aile Büyüklüğü Sütunu Eklenmiş Veri (train_data):")
print(train_data.head())
```

```python
# Cinsiyet (Sex) sütununu dönüştürelim
# 'map' metodu ile 'female' değerlerini 0'a, 'male' değerlerini 1'e dönüştürürüz
train_data['Sex'] = train_data['Sex'].map({'female': 0, 'male': 1})
test_data['Sex'] = test_data['Sex'].map({'female': 0, 'male': 1})

# Biniş Limanı (Embarked) sütununu dönüştürelim
# 'get_dummies' metodu, her bir kategorik değeri (S, C, Q) ayrı bir sütuna dönüştürür
# Bu yöntem, tek bir sütunu 0 ve 1'lerden oluşan birden fazla yeni sütun haline getirir
train_data = pd.get_dummies(train_data, columns=['Embarked'], prefix='Embarked', drop_first=True)
test_data = pd.get_dummies(test_data, columns=['Embarked'], prefix='Embarked', drop_first=True)

# Verinin son halini tekrar kontrol edelim
print("Son Durumdaki Eğitim Verisi:")
print(train_data.head())
```

```python
# 'Survived' sütununu hedef değişken (y) olarak ayırırız
y = train_data['Survived']

# 'Survived' ve 'PassengerId' sütunlarını eğitim verisinden (X) çıkarırız
# 'PassengerId' model için bir anlam ifade etmez ve 'Survived' zaten hedefimizdir
X = train_data.drop(['Survived', 'PassengerId'], axis=1)

# Test verisinden (X_test) sadece 'PassengerId' sütununu çıkarırız
X_test = test_data.drop('PassengerId', axis=1)
```

```python
# Gerekli kütüphaneleri içe aktarırız
from sklearn.linear_model import LogisticRegression

# Modeli oluştururuz
# 'random_state=1' parametresi, her çalıştırdığında aynı sonucu alman için önemlidir
model = LogisticRegression(random_state=1)

# Modeli, eğitim verilerimizi (X) ve hedef değişkenimizi (y) kullanarak eğitiriz
model.fit(X, y)
```

```python
# Gerekli kütüphaneyi içe aktarırız
from sklearn.preprocessing import StandardScaler

# Ölçeklendirme aracımızı (StandardScaler) oluştururuz
scaler = StandardScaler()

# Ölçeklendirme yapılacak sütunları belirleriz
# 'PassengerId', 'Survived', 'Age' ve 'Fare' sütunlarını dahil ederiz
# Pclass gibi kategorik verileri burada bırakırız, çünkü zaten 1, 2, 3 gibi sayısal değerler
columns_to_scale = ['Age', 'Fare']

# Eğitim verisindeki (X) ilgili sütunları ölçeklendiririz
# 'fit_transform()' metodu ile hem öğrenme hem de dönüştürme işlemini tek adımda yaparız
X[columns_to_scale] = scaler.fit_transform(X[columns_to_scale])

# Test verisindeki (X_test) ilgili sütunları da ölçeklendiririz
# Burada sadece 'transform()' kullanırız, çünkü ölçeklendirme kurallarını eğitim verisinden öğrendik
X_test[columns_to_scale] = scaler.transform(X_test[columns_to_scale])

# Verinin son halini görmek için ilk 5 satırını yazdırırız
print("Ölçeklendirilmiş Veri (X):")
print(X.head())
```

```python
# Gerekli kütüphaneyi tekrar içe aktarırız
from sklearn.linear_model import LogisticRegression

# Modeli tekrar oluştururuz ve 'max_iter' değerini 1000'e çıkarırız
# Bu, modelin en iyi çözümü bulmak için daha fazla deneme yapmasına izin verir
model = LogisticRegression(random_state=1, max_iter=1000)

# Modeli, artık ölçeklendirilmiş olan eğitim verileri (X) ve hedef değişkenimiz (y) ile eğitiriz
model.fit(X, y)

print("Model başarıyla eğitildi!")
```

```python
# Eğittiğimiz modeli kullanarak test verisi (X_test) üzerinde tahminler yaparız
# 'predict()' metodu, modelden her bir yolcu için hayatta kalma tahmini (0 veya 1) yapmasını ister
tahminler = model.predict(X_test)

# Tahmin sonuçlarını görmek için ilk 5 tahmini yazdırırız
print("İlk 5 Tahmin:")
print(tahminler[:5])
```

```python
# Pandas kütüphanesini kullanarak yeni bir veri çerçevesi (DataFrame) oluştururuz
# İlk sütun 'PassengerId', test verisindeki yolcu kimlik numaralarından oluşur
# İkinci sütun 'Survived', bizim modelimizin yaptığı tahminlerden oluşur
sonuc_df = pd.DataFrame({'PassengerId': test_data['PassengerId'], 'Survived': tahminler})

# Hazırladığımız veri çerçevesini bir CSV dosyası olarak dışa aktarırız
# 'index=False' parametresi, Pandas'ın ekstra bir indeks sütunu eklemesini engeller
sonuc_df.to_csv('submission.csv', index=False)

print("Kaggle'a gönderebileceğin 'submission.csv' dosyası oluşturuldu!")
```
